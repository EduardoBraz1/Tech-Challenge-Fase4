```python
# Importando a biblioteca Pandas
import pandas as pd

# 1.Carregando o arquivo CSV para um DataFrame
df = pd.read_csv('Obesity.csv')

# 2.Exibindo as primeiras linhas do DataFrame
display(df.head())

# 3.Exibindo as informações gerais do DataFrame
print("\n---Informações da Base de Dados: ---")
df.info()

# 4.Verificar se existem valores nulos
print("\n---Valores Nulos por Coluna: ---")
print(df.isnull().sum())
```

```python
# 1.Arredondar colunas especificas no Dicionario de Dados e transformar o tipo de dado para inteiro (int)
colunas_para_arredondar = ['Age', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']

df[colunas_para_arredondar] = df[colunas_para_arredondar].round(0).astype(int)

# 2.Imprimir colunas para verificar se deu certo o arredondamento
print("--- Primeiras linhas após arredondamento: ---")
display(df[['Age', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']].head())
```

```python
# Tradução do Dicionário de Dados para Português para uso no Power BI
# 1.Dicionario para traduzir o nome das colunas
traducao_colunas = {
    'Gender': 'Genero',
    'Age': 'Idade',
    'Height': 'Altura',
    'Weight': 'Peso',
    'family_history': 'Historico_Familiar',
    'FAVC': 'Consumo_Alta_Caloria',
    'FCVC': 'Frequencia_Vegetais',
    'NCP': 'Numero_Refeicoes',
    'CAEC': 'Consumo_Entre_Refeicoes',
    'SMOKE': 'Fumante',
    'CH2O': 'Consumo_Agua',
    'SCC': 'Monitora_Calorias',
    'FAF': 'Atividade_Fisica',
    'TUE': 'Tempo_Telas',
    'CALC': 'Consumo_Alcool',
    'MTRANS': 'Meio_Transporte',
    'Obesity': 'Nivel_Obesidade'
}

# 2.Dicionarios para traduzir os valores das linhas
traducao_sim_nao = {'yes': 'Sim', 'no': 'Nao'}
traducao_frequencia = {'no': 'Nao', 'Sometimes': 'As vezes', 'Frequently': 'Frequentemente', 'Always': 'Sempre'}

# Apicando as traduções em uma copia da base de dados para não perder a original
df_powerbi = df.copy()

# Traduzindo colunas
df_powerbi.rename(columns=traducao_colunas, inplace=True)

# Traduzindo os valores (linhas)
df_powerbi['Genero'] = df_powerbi['Genero'].map({'Female': 'Feminino', 'Male': 'Masculino'})
df_powerbi['Historico_Familiar'] = df_powerbi['Historico_Familiar'].map(traducao_sim_nao)
df_powerbi['Consumo_Alta_Caloria'] = df_powerbi['Consumo_Alta_Caloria'].map(traducao_sim_nao)
df_powerbi['Fumante'] = df_powerbi['Fumante'].map(traducao_sim_nao)
df_powerbi['Monitora_Calorias'] = df_powerbi['Monitora_Calorias'].map(traducao_sim_nao)

df_powerbi['Consumo_Entre_Refeicoes'] = df_powerbi['Consumo_Entre_Refeicoes'].map(traducao_frequencia)
df_powerbi['Consumo_Alcool'] = df_powerbi['Consumo_Alcool'].map(traducao_frequencia)

# Traduzindo a coluna de numero de refeições (NCP)
df_powerbi['Numero_Refeicoes'] = df_powerbi['Numero_Refeicoes'].map({
    1: '1 refeição',
    2: '2 refeições',
    3: '3 refeições',
    4: '4 refeições ou mais'
})

# Traduzindo a coluna de consumo de vegetais (FCVC)
df_powerbi['Frequencia_Vegetais'] = df_powerbi['Frequencia_Vegetais'].map({
    1: 'Raramente',
    2: 'As vezes',
    3: 'Sempre'
})

# Traduzindo a coluna de consumo de agua (CH2O)
df_powerbi['Consumo_Agua'] = df_powerbi['Consumo_Agua'].map({
    1: '< 1 L/dia',
    2: '1-2 L/dia',
    3: '> 2 L/dia'
})

# Traduzindo a coluna de atividade fisica (FAF)
df_powerbi['Atividade_Fisica'] = df_powerbi['Atividade_Fisica'].map({
    0: 'nenhuma',
    1: '~1-2x/sem',
    2: '~3-4x/sem',
    3: '5x/sem ou mais'
})

# Traduzindo a coluna de tempo de telas (TUE)
df_powerbi['Tempo_Telas'] = df_powerbi['Tempo_Telas'].map({
    0: '~0-2 h/dia',
    1: '~3-5 h/dia',
    2: '> 5 h/dia'
})

#Traduzindo a coluna de meio de transporte (MTRANS)
df_powerbi['Meio_Transporte'] = df_powerbi['Meio_Transporte'].map({
    'Automobile': 'Carro',
    'Motorbike': 'Moto',
    'Bike': 'Bicicleta',
    'Public_Transportation': 'Transporte Publico',
    'Walking': 'A Pe'
})

# Traduzindo a coluna de obesidade (target)
df_powerbi['Nivel_Obesidade'] = df_powerbi['Nivel_Obesidade'].map({
    'Insufficient_Weight': 'Abaixo do Peso',
    'Normal_Weight': 'Peso Normal',
    'Overweight_Level_I': 'Sobrepeso Nivel I',
    'Overweight_Level_II': 'Sobrepeso Nivel II',
    'Obesity_Type_I': 'Obesidade Tipo I',
    'Obesity_Type_II': 'Obesidade Tipo II',
    'Obesity_Type_III': 'Obesidade Tipo III'
})

# 3.Salvar esta base traduzida para uso no Power BI
df_powerbi.to_csv('Obesity_Traduzido_PowerBI.csv', index=False)
print("--- Primeiras lihas da base traduzida: ---")
display(df_powerbi.head())
print("\nArquivo 'Obesity_Traduzido_PowerBI.csv' gerado com sucesso na pasta do projeto!")
```
